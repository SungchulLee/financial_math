삼성전자, 네이버, 카카오 등 우리나라 주식 티커가 주어지면

    1. 야후에서 티커 데이터를 받아와서 

    2. csv파일로 저장하고
    
    3. 저장한 파일을 판다스 데이타프레임으로 불러와
    
    4. 데이터프레임을 리턴하는

함수를 구현하고, 함수가 제대로 구현되었는지 확인하라.

To install yfinance, run the following in terminal:
```
$ pip install yfinance
```
    
[yfinance](https://pypi.org/project/yfinance/)

````python
def download(ticker):
    
    import yfinance as yf
    import pandas as pd
    
    stock = yf.Ticker(ticker)
    data = stock.history(period='max')
    
    data.to_csv(f'./{ticker}.csv')
    
    df = pd.read_csv(f'./{ticker}.csv', 
                     index_col='Date',
                     parse_dates=True)
    
    return df
````

````python
ticker = 'SPY'
df = download(ticker)
df.head()
````
