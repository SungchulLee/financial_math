```python
import os
import pandas as pd
import yfinance as yf
```

```python
def download(ticker, start=None, end=None):
    if start is None:
        return yf.Ticker(ticker).history(period="max")
    else:
        return yf.Ticker(ticker).history(start=start, end=end)
```

```python
url = 'https://raw.githubusercontent.com/datasets/s-and-p-500-companies/master/data/constituents.csv'
df_tickers = pd.read_csv(url)
df_tickers
```

```python
df_sp500 = pd.DataFrame()
for ticker in df_tickers.Symbol:
    if ticker=='BRK.B':
        ticker = 'BRK-B'
    if ticker=='BF.B':
        ticker = 'BF-B'
    df_ = download(ticker).rename(columns={'Close' : ticker})[[ticker]]
    df_sp500 = df_sp500.join(df_ticker, how='outer')
df_sp500
```

```python
os.makedirs('../data/sp500', exist_ok=True)
for ticker in df_tickers.Symbol:
    if ticker=='BRK.B':
        ticker = 'BRK-B'
    if ticker=='BF.B':
        ticker = 'BF-B'
    df_ticker = download(ticker)
    df_ticker.to_csv(f'../data/sp500/{ticker}.csv')
```
