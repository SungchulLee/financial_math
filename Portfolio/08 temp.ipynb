```python
def make_csv_file_path(ticker, data_dir="../data/dow30"): 
    """
    For a given ticker and data_dir, make a csv file path where a csv file is.
    
    Input
    ticker: string
    data_dir: string
    
    Output
    csv_file_path: string
    """ 
    csv_file_path = os.path.join(data_dir, "{}.csv".format(ticker))  
    return csv_file_path 


if __name__ == "__main__":
    ticker = "WMT"
    data_dir = "../data/dow30"
    
    csv_file_path = make_csv_file_path(ticker, data_dir)
    
    print(csv_file_path)    
```

```python
def update_ticker_list_to_include_benchmark_ticker(ticker_list, benchmark_ticker): 
    """
    Update ticker list to include benchmark ticker.
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    
    Output
    ticker_list: list of strings
    """
    if benchmark_ticker not in ticker_list:  
        ticker_list.insert(0, benchmark_ticker) 
    return ticker_list


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    
    ticker_list = update_ticker_list_to_include_benchmark_ticker(ticker_list, benchmark_ticker)
    
    print(ticker_list)  
```

```python
def get_data_and_updated_ticker_list(ticker_list, benchmark_ticker, 
                                     start="2017-01-01", end="2017-12-31", 
                                     data_dir="../data/dow30"): 
    """
    Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    
    Output
    df_open: pd.DataFrame
    df_high: pd.DataFrame
    df_low: pd.DataFrame
    df_close: pd.DataFrame
    df_adj_close: pd.DataFrame
    df_volume: pd.DataFrame
    ticker_list: list of strings
    """ 
    ticker_list = update_ticker_list_to_include_benchmark_ticker(ticker_list, benchmark_ticker)
      
    dates = pd.date_range(start, end)
    
    for data_type in ["Open", "High", "Low", "Close", "Adj Close", "Volume"]:
        
        df = pd.DataFrame(index=dates) 
        for ticker in ticker_list: 
            csv_file_path = make_csv_file_path(ticker, data_dir) 
            df_temp = pd.read_csv(csv_file_path,
                                  index_col="Date",
                                  parse_dates=True, 
                                  usecols=["Date", data_type], 
                                  na_values=["null"]) 
            df_temp = df_temp.rename(columns={data_type: ticker})
            df = df.join(df_temp) 

            if ticker == benchmark_ticker:  
                df = df.dropna(subset=[benchmark_ticker])
                
        if data_type == "Open":
            df_open = df
        if data_type == "High":
            df_high = df   
        if data_type == "Low":
            df_low = df    
        if data_type == "Close":
            df_close = df
        if data_type == "Adj Close":
            df_adj_close = df
        if data_type == "Volume":
            df_volume = df   
             
    return df_open, df_high, df_low, df_close, df_adj_close, df_volume, ticker_list


def run_test_code_1(ticker_list, benchmark_ticker, start, end, data_dir):
    """
    Run test code 1 which do the following:
    1. Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    2. Print out open, high, low, close, adjusted close, volumn, 
    and ticker_list, which has been updated to include benchmark ticker.  
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    """   
    df_open, df_high, df_low, df_close, df_adj_close, df_volume, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    print("Open ========")
    print(df_open.head())
    print()
    print("High ========")
    print(df_high.head()) 
    print()
    print("Low ========")
    print(df_low.head()) 
    print()
    print("Close ========")
    print(df_close.head()) 
    print()
    print("Adj Close ========")
    print(df_adj_close.head()) 
    print()
    print("Volume ========")
    print(df_volume.head()) 
    print()
    print("Ticker List ========")
    print(ticker_list) 


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-01-07"
    data_dir = "../data/dow30"
    
    run_test_code_1(ticker_list, benchmark_ticker, start, end, data_dir)   
```

```python
def compute_daily_return(df_adj_close):
    """
    Compute daily return from daily adj close.
    
    Input
    df_adj_close: pd.DataFrame
    
    Output
    df_daily_return: pd.DataFrame
    """
    df_daily_return = df_adj_close.copy() 
    df_daily_return = (df_adj_close / df_adj_close.shift(1)) - 1 
    df_daily_return.iloc[0,:] = 0 # remove NaN at the begining of df_daily_return   
    return df_daily_return


def run_test_code_2(ticker_list, benchmark_ticker, start, end, data_dir):
    """
    Run test code 2 which do the following:
    1. Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    2. Compute daily return from daily adj close. 
    3. Print adjusted close and daily return.  
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    """ 
    _, _, _, _, df_adj_close, _, _ = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    df_daily_return = compute_daily_return(df_adj_close)
    
    print("Adj Close ========")
    print(df_adj_close) 
    print()
    print("Daily Return ========")
    print(df_daily_return) 


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-01-07"
    data_dir = "../data/dow30"
    
    run_test_code_2(ticker_list, benchmark_ticker, start, end, data_dir) 
```

```python
def compute_expected_return(df_daily_return):
    """
    Compute expected return from daily return.
    
    Input
    df_daily_return: pd.DataFrame
    
    Output
    ps_expected_return: pd.Series
    """
    ps_expected_return = 252 * df_daily_return.mean() 
    return ps_expected_return

def compute_covariance_matrix_of_return(df_daily_return):
    """
    Compute covariance matrix of return from daily return.
    
    Input
    df_daily_return: pd.DataFrame
    
    Output
    df_covariance_matrix_of_return: pd.DataFrame
    """
    df_covariance_matrix_of_return = 252 * df_daily_return.cov() 
    return df_covariance_matrix_of_return

def compute_volatility(df_daily_return):
    """
    Compute volatility from daily return.
    
    Input
    df_daily_return: pd.DataFrame
    
    Output
    ps_volatility: pd.Series
    """
    df = compute_covariance_matrix_of_return(df_daily_return)
    ps_volatility = pd.Series(np.sqrt(np.diag(df)), index=df.columns)
    return ps_volatility


def run_test_code_3(ticker_list, benchmark_ticker, start, end, data_dir):
    """
    Run test code 3 which do the following:
    1. Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    2. Compute daily return from daily adj close. 
    3. Compute expected return from daily return.
    4. Compute covariance matrix of return from daily return.
    5. Compute volatility from daily return.
    6. Print daily return, expected return, covariance matrix, and volatility.  
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    """  
    _, _, _, _, df_adj_close, _, _ = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    df_daily_return = compute_daily_return(df_adj_close)
    
    ps_expected_return = compute_expected_return(df_daily_return)
    df_covariance_matrix_of_return = compute_covariance_matrix_of_return(df_daily_return)
    ps_volatility = compute_volatility(df_daily_return)
    
    print("Daily Return ========")
    print(df_daily_return.head())
    print()
    print("Expected Return ========")
    print(ps_expected_return) 
    print()
    print("Covariance Matrix of Return ========")
    print(df_covariance_matrix_of_return)
    print()
    print("Volatility ========")
    print(ps_volatility)


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    run_test_code_3(ticker_list, benchmark_ticker, start, end, data_dir) 
```

```python
def plot_risk_and_expected_return(ps_volatility, ps_expected_return, ticker_list, alpha=0.3):
    """
    Plot risk and expected return.
    
    Input
    ps_volatility: pd.Series
    ps_expected_return: pd.Series
    alpha: float
    """
    fig, ax = plt.subplots()
     
    for i, ticker in enumerate(ticker_list):
        ax.scatter(ps_volatility[i], ps_expected_return[i], alpha=alpha)
        ax.annotate(ticker, (ps_volatility[i],ps_expected_return[i]))
    
    plt.show()
    
    
def run_test_code_4(ticker_list, benchmark_ticker, start, end, data_dir):
    """
    Run test code 4 which do the following:
    1. Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    2. Compute daily return from daily adj close. 
    3. Compute expected return from daily return.
    4. Compute volatility from daily return.
    5. Plot risk and expected return.
    6. Print expected return and volatility.  
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    """ 
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    df_daily_return = compute_daily_return(df_adj_close)
    
    ps_expected_return = compute_expected_return(df_daily_return)
    ps_volatility = compute_volatility(df_daily_return)
    
    plot_risk_and_expected_return(ps_volatility, ps_expected_return, ticker_list)
    
    print("Expected Return ========")
    print(ps_expected_return) 
    print()
    print("Volatility ========")
    print(ps_volatility)
    

if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    run_test_code_4(ticker_list, benchmark_ticker, start, end, data_dir) 
```

```python
def normalize_data(df):
    """
    Normalize data using the first row of data.
    
    Input
    df: pd.DataFrame
    
    Output
    df_normalized: pd.DataFrame
    """
    df_normalized = df / df.iloc[0,:]
    return df_normalized


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-01-07"
    data_dir = "../data/dow30"
    
    _, _, _, _, df_adj_close, _, _ = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    df_daily_return = compute_daily_return(df_adj_close)
    
    df_normalized_adj_close = normalize_data(df_adj_close)
    
    print("Adj Close ========")
    print(df_adj_close.head()) 
    print()
    print("Normalized Adj Close ========")
    print(df_normalized_adj_close.head()) 
```

```python
def get_portfolio_adj_close(df_adj_close, portfolio_weight, portfolio_name):
    """
    Get portfolio adj close from individual asset adj close and portfolio weight.
    
    Input
    df_adj_close: pd.DataFrame
    portfolio_weight: list of floats
    portfolio_name: string
    
    Output
    df_portfolio_adj_close: pd.DataFrame
    """
    df_normalized_adj_close = normalize_data(df_adj_close) 
    df_portfolio_adj_close = (portfolio_weight * df_normalized_adj_close).sum(axis=1) 
    df_portfolio_adj_close = pd.DataFrame(df_portfolio_adj_close, columns=[portfolio_name])
    return df_portfolio_adj_close


def run_test_code_5(ticker_list, benchmark_ticker, start, end, data_dir, portfolio_weight, portfolio_name):
    """
    Run test code 5 which do the following:
    1. Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    2. Get portfolio adj close from individual asset adj close and portfolio weight.
    3. Print head of adj close and portfolio adj close. 
    4. Join adj close and portfolio adj close and make appended adj close. 
    5. Print head of appended adj close. 
    6. Compute appended daily return.
    7. Compute expected return from appended daily return.
    8. Compute volatility from appended daily return.
    9. Plot risk and expected return.
    10. Print head of expected return and volatility.   
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    portfolio_weight: list of floats
    portfolio_name: string
    """ 
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)

    df_portfolio_adj_close = get_portfolio_adj_close(df_adj_close, portfolio_weight, portfolio_name)
    
    print("Adj Close ========")
    print(df_adj_close.head()) 
    print()   
    print("Portfolio Adj Close ========")
    print(df_portfolio_adj_close.head())
    print() 
    
    df_appended_adj_close = df_adj_close.join(df_portfolio_adj_close)
      
    print("Appended Adj Close  ========")
    print(df_appended_adj_close.head()) 
    
    df_appended_daily_return = compute_daily_return(df_appended_adj_close)
    
    ps_appended_expected_return = compute_expected_return(df_appended_daily_return) 
    ps_appended_volatility = compute_volatility(df_appended_daily_return)
    ticker_list.append(portfolio_name)
    
    plot_risk_and_expected_return(ps_appended_volatility, ps_appended_expected_return, ticker_list)
    
    print("Appended Expected Return ========")
    print(ps_appended_expected_return.head()) 
    print()
    print("Appended Volatility ========")
    print(ps_appended_volatility.head())


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    portfolio_weight = [0.6, 0.2, 0.2]
    portfolio_name = "622"
    
    run_test_code_5(ticker_list, benchmark_ticker, start, end, data_dir, portfolio_weight, portfolio_name)    
```

```python
def get_random_portfolio_weight(n_columns_in_data, seed=None):
    """
    Get random portfolio weight from individual asset adj close.
    
    Input
    n_columns_in_data: integer
    
    Output
    random_portfolio_weight: list
    """
    if seed != None:
        np.random.seed(seed)
    
    random_portfolio_weight = np.random.uniform(0,1,n_columns_in_data)
    random_portfolio_weight = list(random_portfolio_weight / random_portfolio_weight.sum())
    
    return random_portfolio_weight

if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-01-31"
    data_dir = "../data/dow30"
    
    portfolio_weight = get_random_portfolio_weight(len(ticker_list)+1, seed=0)
    portfolio_name = "random"

    run_test_code_5(ticker_list, benchmark_ticker, start, end, data_dir, portfolio_weight, portfolio_name) 
```

```python
def get_GMVP_portfolio_weight(df_adj_close):
    """
    Get GMVP portfolio weight from individual asset adj close.
    
    Input
    df_adj_close: pd.DataFrame
    
    Output
    GMVP_portfolio_weight: list
    """
    df_daily_return = compute_daily_return(df_adj_close)
    
    ps_expected_return = compute_expected_return(df_daily_return)
    df_covariance_matrix_of_return = compute_covariance_matrix_of_return(df_daily_return)

    A = np.array(df_covariance_matrix_of_return)
    b = np.ones_like(np.array(ps_expected_return).reshape((len(ps_expected_return),1)))
    try:
        GMVP_portfolio_weight, _, _, _ = np.linalg.lstsq(A, b, rcond=None)
    except:
        GMVP_portfolio_weight, _, _, _ = np.linalg.lstsq(A, b)
    GMVP_portfolio_weight = (GMVP_portfolio_weight / GMVP_portfolio_weight.sum()).reshape(len(df_adj_close.columns))
    GMVP_portfolio_weight = list(GMVP_portfolio_weight)
    
    return GMVP_portfolio_weight


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-01-31"
    data_dir = "../data/dow30"
    
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(ticker_list, benchmark_ticker, start, end, data_dir)
    portfolio_weight = get_GMVP_portfolio_weight(df_adj_close)
    portfolio_name = "GMVP"

    run_test_code_5(ticker_list, benchmark_ticker, start, end, data_dir, portfolio_weight, portfolio_name) 
```

```python
def get_market_portfolio_weight(df_adj_close, interest_rate):
    """
    Get market portfolio weight from individual asset adj close and risk free interest rate.
    
    Input
    df_adj_close: pd.DataFrame
    interest_rate: float
    
    Output
    market_portfolio_weight: list
    """
    df_daily_return = compute_daily_return(df_adj_close)
    
    ps_expected_return = compute_expected_return(df_daily_return)
    df_covariance_matrix_of_return = compute_covariance_matrix_of_return(df_daily_return)

    A = np.array(df_covariance_matrix_of_return)
    b = np.array(ps_expected_return).reshape((len(ps_expected_return),1)) - interest_rate
    try:
        m_portfolio_weight, _, _, _ = np.linalg.lstsq(A, b, rcond=None)
    except:
        m_portfolio_weight, _, _, _ = np.linalg.lstsq(A, b)
    market_portfolio_weight = (m_portfolio_weight / m_portfolio_weight.sum()).reshape(len(df_adj_close.columns))
    market_portfolio_weight = list(market_portfolio_weight)
    
    return market_portfolio_weight


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    interest_rate = 0.03
    portfolio_weight = get_market_portfolio_weight(df_adj_close, interest_rate)
    portfolio_name = "market"
    
    run_test_code_5(ticker_list, benchmark_ticker, start, end, data_dir, portfolio_weight, portfolio_name) 
```

```python
def plot_risk_return_and_efficient_frontier(df_adj_close, df_GMVP_adj_close, df_market_adj_close, alpha=0.3):
    """
    Plot risk, return, and efficient frontier.
    
    Input
    df_adj_close: pd.DataFrame 
    df_GMVP_adj_close: pd.DataFrame
    df_market_adj_close: pd.DataFrame
    alpha: string
    """
    df = df_adj_close
    df_daily_return = compute_daily_return(df) 
    mu_asset = compute_expected_return(df_daily_return)
    sigma_asset = compute_volatility(df_daily_return)
    
    df = df_GMVP_adj_close.join(df_market_adj_close)
    df_daily_return = compute_daily_return(df)
    mu_GMVP_market = compute_expected_return(df_daily_return)
    Sigma_GMVP_market = compute_covariance_matrix_of_return(df_daily_return)
    
    al = np.linspace(0,1)
    mu_efficient_frontier = al*mu_GMVP_market.iloc[0]+(1-al)*mu_GMVP_market.iloc[1]
    sigma_efficient_frontier = np.sqrt(al**2*Sigma_GMVP_market.iloc[0,0]
                                       + (1-al)**2*Sigma_GMVP_market.iloc[1,1] 
                                       + 2*al*(1-al)*Sigma_GMVP_market.iloc[0,1])
    
    fig, ax = plt.subplots()
     
    for i, ticker in enumerate(df_adj_close.columns):
        ax.scatter(sigma_asset[i], mu_asset[i], alpha=alpha)
        ax.annotate(ticker, (sigma_asset[i],mu_asset[i]))
        
    ax.plot(sigma_efficient_frontier, mu_efficient_frontier)
    ax.annotate("market", (sigma_efficient_frontier[0],mu_efficient_frontier[0]))
    ax.annotate("GMVP", (sigma_efficient_frontier[-1],mu_efficient_frontier[-1]))
    
    plt.show()
    
    
def run_test_code_6(ticker_list, benchmark_ticker, start, end, data_dir, interest_rate):
    """
    Run test code 6 which do the following:
    1. Get (open, high, low, close, adjusted close, volumn) stock data of given tickers and benchmark 
    from CSV files in the data folder, and updated ticker list.
    2. Get GMVP portfolio weight and its adj close.
    3. Get market portfolio weight and its adj close.
    4. Plot risk, return, and efficient frontier.  
    
    Input
    ticker_list: list of strings
    benchmark_ticker: string
    start: string
    end: string
    data_dir: string
    interest_rate: float
    """ 
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    portfolio_weight = get_GMVP_portfolio_weight(df_adj_close)
    portfolio_name = "GMVP"
    df_GMVP_adj_close = get_portfolio_adj_close(df_adj_close, portfolio_weight, portfolio_name)
    
    portfolio_weight = get_market_portfolio_weight(df_adj_close, interest_rate)
    portfolio_name = "market"
    df_market_adj_close = get_portfolio_adj_close(df_adj_close, portfolio_weight, portfolio_name)
    
    plot_risk_return_and_efficient_frontier(df_adj_close, df_GMVP_adj_close, df_market_adj_close)


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    interest_rate = 0.03

    run_test_code_6(ticker_list, benchmark_ticker, start, end, data_dir, interest_rate)
```

```python
def get_train_start_train_end_test_start_test_end(df_adj_close, train_period_in_month):
    """
    Get train_start, train_end, test_start, test_end.
    
    Input
    df_adj_close: pd.DataFrame 
    train_period_in_month: int
    
    Output
    train_start: list of datetime stamps 
    train_end: list of datetime stamps  
    test_start: list of datetime stamps  
    test_end: list of datetime stamps 
    """
    # first date of month
    for i, datetimeobject in enumerate(df_adj_close.index):
        if i == 0:
            temp = [datetimeobject]
        elif datetimeobject.month != datetimeobject_previous.month: 
                temp.append(datetimeobject)
        datetimeobject_previous = datetimeobject
    first_date_of_month = temp
    
    # last date of month
    for i, datetimeobject in enumerate(df_adj_close.index):
        if i == 0:
            temp = []
        elif datetimeobject.month != datetimeobject_previous.month: 
                temp.append(datetimeobject_previous)
        elif i == len(df_adj_close.index)-1: 
                temp.append(datetimeobject)
        datetimeobject_previous = datetimeobject 
    last_date_of_month = temp
    
    train_start = first_date_of_month[0:-train_period_in_month] 
    train_end = last_date_of_month[train_period_in_month-1:-1]
    test_start = first_date_of_month[train_period_in_month:] 
    test_end = last_date_of_month[train_period_in_month:]  

    return train_start, train_end, test_start, test_end


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    train_period_in_month = 3
    
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    train_start, train_end, test_start, test_end = get_train_start_train_end_test_start_test_end(
        df_adj_close, train_period_in_month)

    print(train_start) 
    print(train_end) 
    print(test_start) 
    print(test_end)
```

```python
def get_train_and_test_data(df_adj_close, train_start, train_end, test_start, test_end):
    """
    Plot risk, return, and efficient frontier.
    
    Input
    df_adj_close: pd.DataFrame 
    train_start: string 
    train_end: string
    test_start: string
    test_end: string
    
    Output
    df_train_adj_close: pd.DataFrame 
    df_test_adj_close: pd.DataFrame 
    """
    train_dates = pd.date_range(train_start, train_end) 
    test_dates = pd.date_range(test_start, test_end)
    
    df_train_adj_close = df_adj_close.loc[train_dates,:]
    df_test_adj_close = df_adj_close.loc[test_dates,:] 
    
    df_train_adj_close = df_train_adj_close.dropna(subset=[df_train_adj_close.columns[0]])
    df_test_adj_close = df_test_adj_close.dropna(subset=[df_test_adj_close.columns[0]])
    return df_train_adj_close, df_test_adj_close


if __name__ == "__main__":
    ticker_list = ["AAPL", "WMT"]
    benchmark_ticker = "SPY"
    start, end = "2017-01-01", "2017-12-31"
    data_dir = "../data/dow30"
    
    train_period_in_month = 3
    
    _, _, _, _, df_adj_close, _, ticker_list = get_data_and_updated_ticker_list(
        ticker_list, benchmark_ticker, start, end, data_dir)
    
    train_start, train_end, test_start, test_end = get_train_start_train_end_test_start_test_end(
        df_adj_close, train_period_in_month)
    
    for i in range(len(train_start)): 
        df_train_adj_close, df_test_adj_close = get_train_and_test_data(
            df_adj_close, train_start[i], train_end[i], test_start[i], test_end[i])
        print(df_train_adj_close.head(2))
        print(df_train_adj_close.tail(2))
        print(df_test_adj_close.head(2))
        print(df_test_adj_close.tail(2))
```
