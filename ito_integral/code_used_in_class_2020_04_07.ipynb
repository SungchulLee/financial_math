```python
import numpy as np; np.random.seed(1)
import matplotlib.pyplot as plt 

T = 10 
n = 12*21
n_sim = 5
t = np.linspace(0, T, int(n*T)+1)
dt = t[1] - t[0]

zero_one = np.random.binomial(1, 0.5, size=(int(n*T), n_sim))
pm_one = 2 * zero_one - 1
db = np.sqrt(dt) * pm_one 
b = np.concatenate([np.zeros((1, n_sim)), np.cumsum(db, axis=0)], axis=0)
if 0:
    n_share = b[:-1]
elif 0:
    n_share = t[:-1].reshape((-1,1))  
elif 1:
    n_share = b[:-1] * t[:-1].reshape((-1,1))     
dpl = n_share * db
pl = np.concatenate([np.zeros((1, n_sim)), np.cumsum(dpl, axis=0)], axis=0) 

for i in range(n_sim):
    plt.plot(t, pl[:,i])
    plt.grid()
plt.show()
```
