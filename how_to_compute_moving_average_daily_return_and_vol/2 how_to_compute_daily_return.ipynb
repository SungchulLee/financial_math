```python
import pandas as pd
data = pd.read_csv('../data/etf/spy.csv', parse_dates=True, index_col=0)
```

$$
\log\frac{S_t}{S_{t-1}}
$$

```python
import numpy as np
data['Return_C'] = np.log(data['Close']/data['Close'].shift(1)) # 연속 데일리 리턴
data['Return_D'] = data['Close'].pct_change() # 이산 데일리 리턴
```

```python
data
```

```python
import matplotlib.pyplot as plt
data[['Return_C', 'Return_D']].plot(x='Return_C', 
                                    y='Return_D', 
                                    kind='scatter')
plt.plot(data['Return_C'].values, 
         data['Return_C'].values, 
         alpha=0.2,
         color='red',
         linestyle='--')
plt.show()
```
