```python
import pandas as pd
data = pd.read_csv('../data/etf/spy.csv', parse_dates=True, index_col=0)
data['Return'] = data['Close'].pct_change() # 이산 데일리 리턴
```

```python
data
```

```python
import numpy as np
data['42d-Vol'] = data['Return'].rolling(42).std() * (np.sqrt(12*21))
```

```python
data
```

```python
data[['Close','42d-Vol']].plot(subplots=True, style=['b-o','r--'])
```
