```python
# PDF of N(0,I)

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # needed for ax = fig.gca(projection='3d')
from matplotlib.image import cm # needed for cmap=cm.PiYG

x = np.linspace(-3., 3.)
y = np.linspace(-3., 3.)
X, Y = np.meshgrid(x, y)

F = np.exp(-(X**2+Y**2)/2) / (2*np.pi) # PDF of N(0,I)

fig = plt.figure()
ax = fig.gca(projection='3d') 

ax.plot_surface(X, Y, F, \
                rstride=1, cstride=1, \
                cmap=cm.PiYG, linewidth=0, antialiased=False)

ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
# Linear Approximation

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # needed for ax = fig.gca(projection='3d')
from matplotlib.image import cm # needed for cmap=cm.PiYG

x = np.linspace(0., 2.)
b = np.linspace(1., 3.)
X, B = np.meshgrid(x, b)

F = np.exp(X-1) + (X-1)**2 + (B-2)**2 # Original function
G = X # Linear approximation

fig = plt.figure()
ax = fig.gca(projection='3d') 

ax.plot_surface(X, B, F, \
                rstride=1, cstride=1, \
                cmap=cm.PiYG, linewidth=0, antialiased=False)
ax.plot_surface(X, B, G, \
                rstride=1, cstride=1, \
                cmap=cm.PiYG, linewidth=0, antialiased=False)

ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```

```python
# Quadratic Approximation

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # needed for ax = fig.gca(projection='3d')
from matplotlib.image import cm # needed for cmap=cm.PiYG

x = np.linspace(0., 2.)
b = np.linspace(1., 3.)
X, B = np.meshgrid(x, b)

F = np.exp(X-1) + (X-1)**2 + (B-2)**2 # Original function
# G = X # Linear approximation
H = X + 1.5*(X-1)**2 + (B-2)**2 # Quadratic approximation

fig = plt.figure()
ax = fig.gca(projection='3d') 

ax.plot_surface(X, B, F, \
                rstride=1, cstride=1, \
                cmap=cm.PiYG, linewidth=0, antialiased=False)
ax.plot_surface(X, B, H, \
                rstride=1, cstride=1, \
                cmap=cm.PiYG, linewidth=0, antialiased=False)

ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```
