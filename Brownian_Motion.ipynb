```python
import numpy as np
import matplotlib.pyplot as plt

# set parameters ###############################################################
num_steps_per_year = 12*21
num_years = 10
# set parameters ###############################################################

np.random.seed(337)

# fair coins
u = np.random.uniform(0, 1, (num_steps_per_year*num_years, ))
c = np.ones_like(u)
c[u < 0.5] = -1

# Brownian motion
b = np.cumsum(c) / np.sqrt(num_steps_per_year)
# https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.insert.html
b = np.insert(b, 0, 0)

# time
t = np.linspace(0, num_years, num_steps_per_year*num_years + 1)   

# plot of Brownian motion 
plt.plot(t,b)
plt.title('Brownian motion')
plt.xlabel('time')
plt.ylabel('$B_t$')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt

# set parameters ###############################################################
num_steps_per_year = 12*21
num_years = 10
num_simulation = 10000
# set parameters ###############################################################

# fair coins
u = np.random.uniform(0, 1, (num_simulation, num_steps_per_year*num_years))
c = np.ones_like(u) 
c[u < 0.5] = -1

# Brownian motion
b = np.cumsum(c, axis=1) / np.sqrt(num_steps_per_year)
# https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.insert.html
b = np.insert(b, 0, 0, axis=1) 

# plot of Brownian motion 
plt.hist(b[:,-1], bins=20)
plt.title('Histogram of Brownian motion')
plt.show()
```
