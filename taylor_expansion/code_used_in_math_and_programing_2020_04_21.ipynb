```python
def f(x):
    return 3 * x**3 + 2 * x**2 + x + 7

def f_prime(x):
    return 9 * x**2 + 4 * x + 1

def f_double_prime(x):
    return 18 * x + 4
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
y = f(x)
plt.plot(x, y, '#303033', label='original function')

x0 = 2.
y0 = f(x0)
plt.plot([x0, x0], [y0, y0], 'or')

y_line = f_prime(x0) * (x - x0) + y0 
plt.plot(x, y_line, 'r', label='tangent line')

plt.legend()
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
y = f(x)
plt.plot(x, y, '#303033', label='original function')

x0 = 2.
y0 = f(x0)
plt.plot([x0, x0], [y0, y0], 'or')

y_line = f_prime(x0) * (x - x0) + y0 
plt.plot(x, y_line, '#35e60e', label='tangent line')

y_quadratic = y0 + f_prime(x0)*(x-x0) + (f_double_prime(x0)/2)*(x-x0)**2 
plt.plot(x, y_quadratic, 'r', label='quadratic approximation')

plt.legend()
plt.show()
```

```python
def F(X, Y):
    return np.exp(X-1)*(Y-1)**3 + (X-1)**2 + (Y-2)**2 + (X-1)*(Y-2)

def F_X(X, Y):
    return np.exp(X-1)*(Y-1)**3 + 2*(X-1) + (Y-2)

def F_Y(X, Y):
    return np.exp(X-1)*3*(Y-1)**2 + 2*(Y-2) + (X-1)

def F_XX(X, Y):
    return np.exp(X-1)*(Y-1)**3 + 2 

def F_YY(X, Y):
    return np.exp(X-1)*6*(Y-1) + (Y-2) 

def F_XY(X, Y):
    return np.exp(X-1)*3*(Y-1)**2 + 1
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization!

x = np.linspace(0,2,100) 
y = np.linspace(1,3,100) 
X, Y = np.meshgrid(x, y)
Z = F(X, Y) 

fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(X, Y, Z,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)

x0 = 1
y0 = 2
Z_linear = F(x0,y0) + F_X(x0,y0)*(X-x0) + F_Y(x0,y0)*(Y-y0)
ax.plot_surface(X, Y, Z_linear,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization!

x = np.linspace(0,2,100) 
y = np.linspace(1,3,100) 
X, Y = np.meshgrid(x, y)
Z = F(X, Y) 

fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(X, Y, Z,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)

x0 = 1
y0 = 2
Z_linear = F(x0,y0) + F_X(x0,y0)*(X-x0) + F_Y(x0,y0)*(Y-y0)
ax.plot_surface(X, Y, Z_linear,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)

FXX = F_XX(x0,y0)*(X-x0)**2/2 
FYY = F_YY(x0,y0)*(Y-y0)**2/2 
FXY = F_XY(x0,y0)*(X-x0)*(Y-y0)
Z_quadratic = Z_linear + FXX + FYY + FXY
ax.plot_surface(X, Y, Z_quadratic,
                cmap=plt.cm.hot,
                linewidth=0.5,
                antialiased=True)

plt.show()
```
