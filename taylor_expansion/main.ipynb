```python
def f(x):
    return 3 * x**3 + 2 * x**2 + x + 7

def f_prime(x):
    return 9 * x**2 + 4 * x + 1

def f_double_prime(x):
    return 18 * x + 4
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
y = f(x)
plt.plot(x,y)

x0 = 2.
y0 = f(x0)
y_prime_0 = f_prime(x0)
y_tangent = y_prime_0 * (x - x0) + y0  
plt.plot(x,y_tangent)

plt.plot([x0, x0], [y0, y0], 'or')

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 100)
y = f(x)
plt.plot(x,y)

x0 = 2
y0 = f(x0)
y_prime_0 = f_prime(x0)
y_double_prime_0 = f_double_prime(x0)
y_quadratic = (1/2)*y_double_prime_0*(x-x0)**2 + y_prime_0*(x-x0) + y0  
plt.plot(x,y_quadratic)

plt.plot([x0, x0], [y0, y0], 'or')

plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization! 
x = np.linspace(-3,3,100)
y = np.linspace(-3,3,100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2+Y**2)/2) / (2*np.pi)

fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(X, Y, Z,
                rstride=2,
                cstride=2,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization!

t = np.linspace(0,2,100) 
b = np.linspace(1,3,100)                 
T, B = np.meshgrid(t,b)
F = np.exp(T-1) + (T-1)**2 + (B-2)**2 # Original function
G = T # Linear approximation
H = T + 1.5*(T-1)**2 + (B-2)**2 # Quadratic approximation

fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(T, B, F,
                rstride=2,
                cstride=2,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)
ax.plot_surface(T, B, G,
                rstride=2,
                cstride=2,
                cmap=plt.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)
plt.show()
```
