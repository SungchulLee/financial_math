티커가 주어지면

1. 야후에서 티커 데이터를 받아와서 

2. 20일 평균, 40일 평균을 계산하고

3. 최근 60일의 종가, 20일 평균, 40일 평균을 하나의 그림에 그리는

함수를 구현하고, 함수가 제대로 구현되었는지 확인하라.

To install yfinance, run the following in terminal:
```
$ pip install yfinance
```
    
[yfinance](https://pypi.org/project/yfinance/)

````python
def draw(ticker):
    import yfinance as yf
    stock = yf.Ticker(ticker)
    data = stock.history(period='max')
    data['20'] = data.Close.rolling(20).mean()
    data['40'] = data.Close.rolling(40).mean()
    data[['Close', '20', '40']][-60:].plot(title=ticker)
````

````python
ticker = '^VIX' # 변동성 지수
draw(ticker)
````

````python
ticker = 'VIXY' # VIX 추종
draw(ticker)
````

````python
ticker = 'UVXY' # VIX 추종 (x 1.5)
draw(ticker)
````
