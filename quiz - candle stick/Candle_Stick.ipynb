1. 야후에서 티커 데이터를 받아와서 

2. 캔들 스틱을 그리는

함수를 구현하고, 함수가 제대로 구현되었는지 확인하라.

To install yfinance, run the following in terminal:
```
$ pip install yfinance
```
    
[yfinance](https://pypi.org/project/yfinance/)

To install mplfinance, run the following in terminal:
```
$ pip install mplfinance
```

[mplfinance](https://pypi.org/project/mplfinance/)

````python
def draw(ticker):
    
    import yfinance as yf
    
    stock = yf.Ticker(ticker)
    data = stock.history(period='max')
    
    import mplfinance as mpf
    
    #mc = mpf.make_marketcolors(up='tab:green', down='tab:red')
    mc = mpf.make_marketcolors(up='tab:red', down='tab:blue')

    s = mpf.make_mpf_style(marketcolors=mc)

    mpf.plot(data[-60:],
             type="candle",
             title=ticker,
             ylabel='Price ($)',
             figratio=(12,4),
             volume=True,
             ylabel_lower='Shares\nTraded',
             show_nontrading=True,
             style=s)
````

````python
ticker = '^VIX' # 변동성 지수
draw(ticker)
````

````python
ticker = 'VIXY' # VIX 추종
draw(ticker)
````
