```python
import pandas as pd
df_pct_change = pd.read_csv('../data/sp500/Pct_Change.csv', 
                            index_col='Date', 
                            parse_dates=True)
```

```python
df_corr = df_pct_change.corr()
print(df_corr)
```

```python
df_corr.to_csv('../data/sp500/Corr.csv')
df_corr = pd.read_csv('../data/sp500/Corr.csv', index_col=0)
print(df_corr)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

img = ax.imshow(df_corr.values, cmap=plt.cm.RdYlGn)
ax.set_xticklabels(df_corr.columns, rotation=90)
ax.set_yticklabels(df_corr.index)

fig.colorbar(img)

plt.show()
```
