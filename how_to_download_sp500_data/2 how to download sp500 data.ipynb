```python
import pandas as pd
tickers = pd.read_csv('../data/sp500/tickers.csv', usecols=[1])
tickers
```

```python
tickers = list(tickers.iloc[:,0])
tickers
```

```python
import yfinance as yf
#import time
import os
tickers_len = len(tickers)
for count, ticker in enumerate(tickers):
    if not os.path.exists('../data/sp500/{}.csv'.format(ticker)):
        stock = yf.Ticker(ticker)
        data = stock.history(period='max')
        data.to_csv('../data/sp500/{}.csv'.format(ticker))
        print('{}/{} : {} Downloaded'.format(count+1, tickers_len, ticker))
        #time.sleep(0.5)
    else:
        print('{}/{} : {} Downloaded'.format(count+1, tickers_len, ticker))
```
