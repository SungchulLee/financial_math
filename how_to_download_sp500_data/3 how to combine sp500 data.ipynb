```python
import pandas as pd
tickers = pd.read_csv('../data/sp500/tickers.csv', usecols=[1])
tickers
```

```python
tickers = list(tickers.iloc[:,0])
tickers
```

```python
df_close = pd.DataFrame()

for count, ticker in enumerate(tickers):
    
    df = pd.read_csv('../data/sp500/{}.csv'.format(ticker), 
                     index_col='Date',
                     parse_dates=True)
    df.rename(columns={'Close':ticker}, inplace=True)
    
    # 0, or ‘index’ : Drop rows which contain missing values. (Default)
    # 1, or ‘columns’ : Drop columns which contain missing value.
    df.drop(['Open','High','Low','Volume','Dividends','Stock Splits'],
            axis=1,
            inplace=True)
    
    # 0, or ‘index’ : Drop rows which contain missing values. (Default)
    # 1, or ‘columns’ : Drop columns which contain missing value.
    df.dropna(axis=0, inplace=True) # 배당 혹은 주식분할로 생기는 추가적인 가로열 제거
    
    if df_close.empty:
        df_close = df
    else:
        df_close = df_close.join(df, how='outer')
        
    if count % 10 == 0:
        print(count)
```

```python
df_close.to_csv('../data/sp500/Close.csv')
df_close = pd.read_csv('../data/sp500/Close.csv', index_col='Date', parse_dates=True)
```
