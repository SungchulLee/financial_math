```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Call(S,K,T,r,v,d=0):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r-d+0.5*v**2)*T)/(v*np.sqrt(T));
    d2 = d1-v*np.sqrt(T);
    C = S*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2);
    return C
```

```python
def Put(S,K,T,r,v,d=0):
    import numpy as np
    from scipy.stats import norm
    d1 = (np.log(S/K)+(r-d+0.5*v**2)*T)/(v*np.sqrt(T));
    d2 = d1-v*np.sqrt(T);
    P = - S*norm.cdf(-d1) + K*np.exp(-r*T)*norm.cdf(-d2);
    return P
```

```python
S = np.linspace(50,150,100); K = 100; T = 1; r = 0.03; v = 0.2;
S = np.concatenate([S, np.array([K])])
S = np.sort(S)

C_BS = Call(S,K,T,r,v);
C_Payoff = np.maximum(S-K,0);
plt.plot(S,C_BS)
plt.plot(S,C_Payoff,'-r');
plt.grid()
plt.show()

P_BS = Put(S,K,T,r,v); 
P_Payoff = np.maximum(K-S1,0); 
plt.plot(S,P_BS)
plt.plot(S,P_Payoff,'-r');
plt.grid()
plt.show()
```
