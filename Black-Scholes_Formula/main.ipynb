```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def d1(S,K,T,r,sigma):
    return (np.log(S/K)+(r+0.5*sigma**2)*T) / (sigma*np.sqrt(T))

def d2(S,K,T,r,sigma):
    return (np.log(S/K)+(r-0.5*sigma**2)*T) / (sigma*np.sqrt(T))

def call(S,K,T,r,sigma):
    d_1 = d1(S,K,T,r,sigma) 
    d_2 = d2(S,K,T,r,sigma) 
    return S*stats.norm.cdf(d_1) - K*np.exp(-r*T)*stats.norm.cdf(d_2)

def put(S,K,T,r,sigma):
    d_1 = d1(S,K,T,r,sigma) 
    d_2 = d2(S,K,T,r,sigma) 
    return -S*stats.norm.cdf(-d_1) + K*np.exp(-r*T)*stats.norm.cdf(-d_2)

K = 100
T = 1/12.
r = 0.5
sigma = 0.3

S = np.linspace(50,150,100)
call_price = call(S,K,T,r,sigma)
call_payoff_if_exercised_now = np.maximum(S-K,0)

plt.plot(S,call_price,'r')
plt.plot(S,call_payoff_if_exercised_now,'--b')
plt.show()
```
