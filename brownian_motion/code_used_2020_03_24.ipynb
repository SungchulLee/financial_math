```python
import numpy as np
np.random.seed(11)

p = 0.5
T = 10
num_flips = 12*21*T
num_simulations = 5

a = np.random.binomial(1, p, size=(num_flips, num_simulations)) 
b = 2 * a - 1
c = np.cumsum(b, axis=0)
rw = np.concatenate([np.zeros((1, num_simulations)), c], axis=0)
bm = rw / np.sqrt(12*21)

dt = 1 / (12*21)
time = dt * np.arange(num_flips+1)  

%matplotlib inline
import matplotlib.pyplot as plt
for i in range(5):
    plt.plot(time, bm[:,i])
plt.grid()
plt.show()
```
