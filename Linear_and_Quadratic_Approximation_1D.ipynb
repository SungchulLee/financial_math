```python
# Linear approximation

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0., 2.)

f = np.exp(x-1) + (x-1)**2 # Original function
g = x # Linear approximation

plt.plot(1, 1, 'or')
plt.plot(x, f, '-') # Plot of original function 
plt.plot(x, g, '-r') # Plot of inear approximation 
plt.show()
```

```python
# Quadratic approximation

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0., 2.)

f = np.exp(x-1) + (x-1)**2 # Original function
# g = x # Linear approximation
h = x + 1.5*(x-1)**2 # Quadratic approximation

plt.plot(1, 1, 'or')
plt.plot(x, f, '-') # Plot of original function 
# plt.plot(x, g, '-r') # Plot of inear approximation 
plt.plot(x, h, '-r') # Plot of quadratic approximation 
plt.show()
```
