```python
import pandas as pd
import numpy as np
data = pd.read_csv('../data/etf/spy.csv', parse_dates=True, index_col=0)
data['Return'] = data['Close'].pct_change()
data['42d_Vol'] = data['Return'].rolling(42).std()*np.sqrt(12*21)
```

```python
data
```

```python
data[['Close','Return', '42d_Vol']].plot(subplots=True,
                                         style='b-o',
                                         figsize=(8, 5))
```
